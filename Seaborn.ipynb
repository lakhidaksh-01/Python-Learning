```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns


# Step-by-Step: Load a Sample Dataset:

# List all available datasets
print(sns.get_dataset_names())

# Load the 'tips' dataset
df = sns.load_dataset('tips')
print(df.head())
```

```python
# Bar Plot 
sns.barplot(x='day',y='total_bill',data=df)
# sns.barplot(x='day',y='total_bill',data=df,color='r')
plt.show()
```

```python
sns.barplot(x='day',y='total_bill',data=df,palette='rocket')
plt.show()
```

```python
# Count Plot

sns.countplot(x='smoker',data=df,palette='Set1')
plt.show()
```

```python
sns.countplot(y='smoker',data=df,palette='Set2')
plt.show()
```

```python
sns.countplot(x='day',data=df,hue='smoker',palette='Set1')
plt.show()
```

```python
# Hist Plot
sns.histplot(df['tip'],kde=True,color='g')
plt.show()
```

```python
# Heat map
sns.heatmap(df.corr(numeric_only=True),cbar=True,linewidths=0.5)
plt.show()
```

```python
# Scatter Plot
df=sns.load_dataset('iris')
sns.scatterplot(x='sepal_length',y='petal_length',data=df,hue='species',size='species')
plt.show()
```

```python
# lmplot
sns.lmplot(x='petal_length',y='petal_width',data=df,hue='species',markers=["o","*","^"])
plt.show()
```

```python
# Box Plot
sns.boxplot(x='petal_length',y='species',data=df)
plt.show()
```
