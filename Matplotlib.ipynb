```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

# Basics
x=np.array([1,2,3,4,5])
y=x**2
plt.plot(x,y)
plt.show()
```

```python
# From csv Data
df=pd.read_csv("sample2.csv")
x=df["Area"]
y=df["Price"]

plt.plot(x,y)
plt.show()
```

```python
# Customizing line plot
df=pd.read_csv("sample2.csv")
x=df["Area"]
y=df["Price"]

plt.plot(x,y,color='green',marker="o",linestyle="-")
plt.title("Line plot")
plt.xlabel('Area (sq. ft.)')
plt.ylabel('Prices')


plt.show()
```

```python
# Diving into subplots
df=pd.read_csv("sample2.csv")

filtered1 = df[df["Price"] < 300000]
x1 = filtered1["Area"]
y1 = filtered1["Price"]

filtered2 = df[df["Price"] > 300000]
x2 = filtered2["Area"]
y2 = filtered2["Price"]

plt.subplot(1,2,1)
plt.plot(x1,y1)

plt.subplot(1,2,2)
plt.plot(x2,y2)

plt.tight_layout()
plt.show()
```

```python
data={'C++':20,'Java':30,'Python':40} # Dictionary storing course names as keys and values as numbers
course=list(data.keys())
values=list(data.values())

plt.bar(course,values)
plt.show()
```

```python
# Customizing Bar graphs
data={'C++':20,'Java':30,'Python':40}
course=list(data.keys())
values=list(data.values())

plt.bar(course,values,color="orange") # if plt.barh(course,values,color="orange") then horizontal.
plt.xlabel("Programming Languages")
plt.ylabel("Popularity Score")
plt.title("Programming Language Popularity")

plt.show()
```

```python
# Box plot
one=[1,2,3,4,5,6,7,8,9]
two=[1,2,3,4,5,4,3,2,1]
three=[6,7,8,9,8,7,6,5,4]

boxdata=[one,two,three]
plt.boxplot(boxdata)
plt.show()
```

```python
# violin plot
one=[1,2,3,4,5,6,7,8,9]
two=[1,2,3,4,5,4,3,2,1]
three=[6,7,8,9,8,7,6,5,4]

violindata=[one,two,three]
plt.violinplot(violindata,showmeans=True,showmedians=True)
plt.grid(True)
plt.show()
```

```python
months=np.arange(1,13)

# Sales data for different products (random values)
Product_A = [30, 40, 35, 50, 55, 65, 70, 75, 80, 85, 90, 100]
Product_B = [20, 30, 25, 35, 45, 50, 55, 60, 65, 70, 75, 85]
Product_C = [10, 20, 15, 25, 30, 35, 40, 45, 50, 55, 60, 70]

plt.stackplot(months,Product_A,Product_B,Product_C,labels=['Product_A','Product_B','Product_C'],colors=['red', 'blue', 'green'],alpha=0.8)

plt.xlabel('Months')
plt.ylabel('Sales')
plt.title('Monthly Sales of Products')

plt.show()
```
